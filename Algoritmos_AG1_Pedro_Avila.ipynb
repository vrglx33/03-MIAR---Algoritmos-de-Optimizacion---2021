# Operaciones

`T(1) = 1`  

`T(n)= 1 + 2 * T(N-1) = 1 + 2*(1+2*T(N-2))) = 1+2 +2^2 + ... + 2*T^2(0)= 2^n-1`

Como 2^n-1 es 2^n  la complejidad es:
 
`O(2^n)`

```python
#Cambio de monedas - Tcnica voraz
################################################################################
################################################################
def cambio_monedas(cantidad,sistema):
  valor_acumulado = 0
  solution_range = range(len(sistema))
  solucion = [0 for i in range(len(sistema))]
  for i in range(len(sistema)):
    monedas = int((cantidad - valor_acumulado)/sistema[i])
    solucion[i] = monedas
    valor_acumulado = valor_acumulado + sistema[i]*monedas
    if valor_acumulado == cantidad:
      return solucion

  if valor_acumulado != cantidad:
    print("no hay solucion")
  else:
    return solucion
sorted_system = sorted([11,5,1], reverse=True)
print(cambio_monedas(15,sorted_system))
################################################################
```

```python
#N Reinas - Vuelta Atrs()
################################################################################

#Verifica que en la solucin parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solucin tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " est " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:       
      return False
  
    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la solucin al tablero
################################################################
def escribe_solucion(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
def reinas(N,solucion=[],etapa=0): 
  if len(solucion)== 0:
    solucion = [0 for i in range(N)]
  
  for i in range(1,N+1):
    solucion[etapa]=i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print("la solucion es: ", solucion)
        escribe_solucion(solucion)
      else: 
        reinas(N,solucion,etapa+1)
    else:
      #print("no es prometedora")
      None
  solucion[etapa] = 0

reinas(10,[],0)


```
